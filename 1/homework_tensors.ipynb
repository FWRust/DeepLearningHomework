```python
import torch
# Создайте следующие тензоры:
print(torch.rand(3,4))# - Тензор размером 3x4, заполненный случайными числами от 0 до 1
print(torch.zeros(2,3,4))# - Тензор размером 2x3x4, заполненный нулями
print(torch.ones(5,5))# - Тензор размером 5x5, заполненный единицами
print(torch.reshape(torch.arange(0,16),(4,4)))# - Тензор размером 4x4 с числами от 0 до 15 (используйте reshape)
```

```python
A = torch.rand(3,4)# Дано: тензор A размером 3x4 и тензор B размером 4x3
B = torch.rand(4,3)
# Выполните:
print(A.T)# - Транспонирование тензора A
print(A@B)# - Матричное умножение A и B
print(A*B.T)# - Поэлементное умножение A и транспонированного B
print(torch.sum(A))# - Вычислите сумму всех элементов тензора A
```

```python
A = torch.reshape(torch.arange(125),(5,5,5))# Создайте тензор размером 5x5x5
# Извлеките:
print(A[0,0,:])# - Первую строку
print(A[-1,:,-1])# - Последний столбец
print(A[2,1:3,1:3])# - Подматрицу размером 2x2 из центра тензора
print(A[::2,::2,::2])# - Все элементы с четными индексами
```

```python
A = torch.arange(24)# Создайте тензор размером 24 элемента
# Преобразуйте его в формы:
print(A.reshape(2,12))# - 2x12
print(A.reshape(3,8))# - 3x8
print(A.reshape(4,6))# - 4x6
print(A.reshape(2,3,4))# - 2x3x4
print(A.reshape(2,2,2,3))# - 2x2x2x3
```
